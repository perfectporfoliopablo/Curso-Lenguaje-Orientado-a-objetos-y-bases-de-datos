```python
font_algerian={'fontname':'Algerian'}
x=[1,3,5,8,10,12,15,18]
y=[0.3,0.5,0.7,1.10,1.30,1.50,1.60,1.75]

font1={'fontfamily':'fantasy','color':'green','size':15}
font2={'fontfamily':'serif','color':'purple','size':10}
plt.plot(x,y,marker='*')
plt.title('Crecimiento del ser humano',fontdict=font1, loc='left')
plt.xlabel('Edad',fontdict=font2)
plt.ylabel('Altura',fontdict=font2)
plt.grid(color='red',ls=':',lw=1)
plt.show()
```

```python
data=np.random.rand(10,10)
plt.imshow(data,cmap='hot',interpolation='nearest')
plt.xlabel('Venta de productos')
plt.ylabel('Ubicacion de la Tienda')
plt.title('Mpa de calor de las ventas')
plt.colorbar()
plt.show()
```

```python
data={'Espaa':[8,5,6,3,2,4],'Italia':[2,5,2,8,4,3],'Rusia':[0,8,2,6,2,3]}
df=pd.DataFrame(data)
plt.imshow(df,cmap='Blues',interpolation='nearest')
plt.xlabel('Paises')
plt.ylabel('Nivel de satisfaccion')
plt.title('Nivel de satisfaccion de Paises')
plt.colorbar()
plt.show()
```

```python
data={'Pasadena':[1,4,6,3,7,9],'San diego':[2,5,3,8,3,2],'Chula vista':[4,8,1,0,3,6],'San Bernandino':[2,7,1,3,5,9],'San Francisco':[3,2,6,4,5,1],'Los Angeles':[2,7,8,4,3,6]}
df=pd.DataFrame(data)
plt.imshow(df,cmap='Reds',interpolation='nearest')
plt.xlabel('Ciudades')
plt.ylabel('Datos')
plt.xticks(df.index,labels=df.columns,rotation=90)
plt.yticks([0,1,2,3,4,5],labels=['metros cuadrados','puertas','baos','ubicacion','habitaciones','ventanas'])
plt.title('Datos Ciudades')
plt.colorbar()
plt.show()
```

```python
num=10
df=pd.read_excel('/content/California_real_estate.xlsx')
filt=df[['1stFlrSF','2ndFlrSF','TotalBsmtSF','GarageArea','WoodDeckSF']].head(num).T
filt
```

```python
plt.imshow(filt, cmap='Greens', interpolation='nearest')
plt.colorbar()
plt.yticks(range(len(filt.index)),labels= filt.index)
plt.xticks(range(num))
for i in range(filt.shape[0]):
  for j in range(filt.shape[1]):
    plt.text((j-.45),i,(f'{filt.iloc[i,j]}SF'),fontsize=6, color='r',weight='bold')
plt.show()
```
